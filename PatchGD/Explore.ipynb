```python
import torch
import glob
import matplotlib.pyplot as plt
```

```python
l1_grad_1_gpu = [torch.load(f"L1_grad_1gpu_{i}.pt",map_location="cpu") for i in range(5)]
l2_grad_1_gpu = [torch.load(f"L2_grad_1gpu_{i}.pt",map_location="cpu") for i in range(5)]
l2fl_grad_1_gpu = [torch.load(f"L2fl_grad_1gpu_{i}.pt",map_location="cpu") for i in range(5)]
l1_grad_4_gpu = [torch.load(f"L1_grad_4gpu_{i}.pt",map_location="cpu") for i in range(5)]
l2_grad_4_gpu = [torch.load(f"L2_grad_4gpu_{i}.pt",map_location="cpu") for i in range(5)]
l2fl_grad_4_gpu = [torch.load(f"L2fl_grad_4gpu_{i}.pt",map_location="cpu") for i in range(5)]
```

```python
#plotting average gradients 
l1_grad_1_gpu_avg = [torch.mean(tensor) for tensor in l1_grad_1_gpu]
l2_grad_1_gpu_avg = [torch.mean(tensor) for tensor in l2_grad_1_gpu]
l2fl_grad_1_gpu_avg = [torch.mean(tensor) for tensor in l2fl_grad_1_gpu]
l1_grad_4_gpu_avg = [torch.mean(tensor) for tensor in l1_grad_4_gpu]
l2_grad_4_gpu_avg = [torch.mean(tensor) for tensor in l2_grad_4_gpu]
l2fl_grad_4_gpu_avg = [torch.mean(tensor) for tensor in l2fl_grad_4_gpu]
```

```python
l1_grad_4_gpu_avg
```

```python
plt.figure()
plt.plot([1,2,3,4,5],l2_grad_1_gpu_avg,label="L2-1gpu")
plt.plot([1,2,3,4,5],l2_grad_4_gpu_avg,label="L2-4gpu")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot([1,2,3,4,5],l2fl_grad_1_gpu_avg,label="L2fl-1gpu")
plt.plot([1,2,3,4,5],l2fl_grad_4_gpu_avg,label="L2f1-4gpu")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot([1,2,3,4,5],l1_grad_1_gpu_avg,label="L1-1gpu")
plt.plot([1,2,3,4,5],l1_grad_4_gpu_avg,label="L1-4gpu")
plt.legend()
plt.show()
```
